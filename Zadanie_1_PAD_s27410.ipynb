```python
def check_range(x,y,z):
  
  if x in range(y,z):
    print('x jest między y a z')
    
  else:
    print('x NIE jest między y a z')
```

```python
check_range(34, 9, 228)
```

```python
check_range(7, 2, 5)
```

```python
def bool_range(x,y,z):

  if x in range(y,z):
    print(True)
  else:
    print(False)
```

```python
bool_range(7, 5, 20)
```

```python
bool_range(67, 22, 25)
```

```python
def unique_list(my_list):
  
  print(list(set(my_list)))
```

```python
my_list = [1,3,5,6,4,3,2,3,3,4,3,4,5,6,6,4,3,2,12,3,5,63,4,5,3,3,2]

unique_list(my_list)
```

```python
unique_list = [x for i, x in enumerate(my_list) if i == my_list.index(x)]

print(unique_list)
```

```python
def volume_of_sphere(r):

  pi = 3.14
  formula = ((4/3)*pi*(r**3))

  return round(formula, 2) 
```

```python
volume_of_sphere(2)
```

```python
def num_fact(n):
  num = 1
  for number in range(2, n + 1):
        num *= number
  return(num)
```

```python
num_fact(10)
```
