```python
# Zadanie 1

import pandas as pd

df = pd.read_csv('PAD_03_PD.csv', sep = ';')

df

```

```python
df['Country'].value_counts()
```

```python
# Zadanie 2:

df['owned_goods'] = df['owns_car'] + df['owns_TV'] + df['owns_house'] + df['owns_Phone']
df
```

```python
# Zadanie 3:

df_mf = df.groupby('gender')
round(df_mf.aggregate({'owned_goods' : 'mean'}), 2)
```

```python
# Zadanie 4:

ddf = round(df.groupby('Country').aggregate({'owned_goods': 'mean', 'Age': 'min'}), 2)

ddf
```
