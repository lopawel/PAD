```python
# Zadanie 1

import numpy as np
import pandas as pd

path = 'LAB02_numpy/'

data = pd.read_csv(path+'president_heights.csv')
heights = np.array(data['height(cm)'])

meanh = np.mean(heights)
stdh = np.std(heights)
minh = np.min(heights)
maxh = np.max(heights)
perc25 = np.percentile(heights, 25)
medh = np.median(heights)
perc75 = np.percentile(heights, 75)


print("Mean height: {}cm".format(meanh.round(2)))
print("Standard deviation: {}cm".format(stdh.round(2)))
print("Minimum height: {}cm".format(minh.round(2)))
print("Maximum height: {}cm".format(maxh.round(2)))
print("25th percentile: {}cm".format(perc25.round(2)))
print("Median: {}cm".format(medh.round(2)))
print("75th percentile: {}cm".format(perc75.round(2)))
```

```python
# Zadanie 2

import numpy as np
import pandas as pd
path = 'LAB02_numpy/'

data2 = np.genfromtxt(path+'Zadanie_2.csv', delimiter = ';')

arr = np.array(data2)

l2, l3 = np.linalg.eig(arr) 
# l3 = np.linalg.eigvals(arr) 
rarr = np.linalg.inv(arr)


print('Znalezione wektory własne dla macierzy: \n')
print(l2)
print('\nZnalezione wartości własne macierzy: \n')
print(l3)
print('\nMacierz odwrotna z macierzy wyjściowej: \n')
print(rarr)
```

```python
# Zadanie 3

import matplotlib.pyplot as plt
import seaborn; seaborn.set()
import numpy as np
import pandas as pd
path = 'LAB02_numpy/'
%matplotlib inline


rainfall = pd.read_csv(path+'Seattle2014.csv')['PRCP'].values
inches = rainfall / 254.00
inches.shape


plt.hist(inches, 40)
plt.xlabel('inches [inch]')
plt.ylabel('counts')
plt.show()


no_rain = inches[inches == 0].shape[0]
rain_all = inches[inches != 0].shape[0]
heavy_rain = inches[inches > 0.5].shape[0]
little_rain = inches[(inches < 0.2) & (inches != 0)].shape[0]


med_rain = np.median(inches[inches != 0])
med_summ = np.median(inches[172:263])
max_summ = np.max(inches[172:263])
max_non = np.max(np.concatenate((inches[:172], inches[263:])))

        
print('Number days without rain: ' + str(no_rain))
print('Number days with rain: ' + str(little_rain))
print('Days with more than 0.5 inches: ' + str(heavy_rain))
print('Rainy days with < 0.2 inches: ' + str(rain_all))


print('Median precip on rainy days in 2014 (inches): ' + str(round(med_rain, 2)))
print('Median precip on summer days in 2014 (inches):  ' + str(round(med_summ, 2)))
print('Maximum precip on summer days in 2014 (inches): ' + str(round(max_summ, 2)))
print('Maximum precip on non-summer rainy days (inches): ' + str(round(max_non, 2)))


    
   

```

```python
# Zadanie 4

import pandas as pd

A = [0,3,2,5]
B = [0,3,1,4]
a = 4

adds = np.add(A, B)
subs = np.subtract(B, A)
muls = np.multiply(A, a)
skal = np.dot(A, B)
lenv = np.linalg.norm(B)

print(f'Wynik dodawania wektora A i B to: {adds}')
print(f'Wynik odejmowania wektora B od A to: {subs}')
print(f'Wynik mnożenia wektora A przez skalar równy 4 wynosi: {muls}')
print(f'Wynik mnożenia skalarnego obydwu wektorów wynosi: {skal}')
print(f'Długość wektora B wynosi: {round(lenv, 2)}')
```
